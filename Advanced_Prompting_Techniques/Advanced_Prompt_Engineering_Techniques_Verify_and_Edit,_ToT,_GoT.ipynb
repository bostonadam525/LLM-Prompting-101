## 1. Verify and Edit Prompting
* Enhances answer accuracy by using **externals sources** to verify and refine resources.
* Goes beyond technique of **Self-consistency** prompting by using validation from external sources such as Wikipedia or Google.

### Pros - Verify and Edit Prompting
1. Increases LLM accuracy by fact checking initial guesses.
2. Clarifies LLMs thought process for greater transparency.
3. Uses **real-world data** enhancing trust and applicability.
4. Outperforms other methods in complex problem solving (e.g. Chain of Thought, Self-Consistency)

### Cons - Verify and Edit Prompting
1. Most effective for complex questions
   * Less effective on simple/commonsense queries
2. Depends on a reliable verification system to improve answer quality.
3. Risks introducing errors or irrelevant data during edit process!

## 2. Tree of Thoughts Prompting
* Prompt strategy to solve problems by exploring different paths or "thoughts" that lead to a solution.
* Breaks down problem-solving process into smaller parts called **"thought steps"**.
* This is very useful for more complex queries or processes where the LLM needs to break down the steps to find the answer.
* How it works:
1. First step
  * Generates multiple 'thoughts' per step, selecting most promising to solve problems.
  * Classifies 'thoughts' as 'positive' vs. 'negative'.

Screenshot 2024-08-26 at 11.14.46â€¯AM.png

2. Second Step
    * Employs straightforward search tactics to determine which paths to pursue.

3. Third Step
    * Resembles a decision tree, exploring various 'thought branches' for optimal solutions to queries.

#### Differences betwen Tree of Thoughts and Chain-of-Thoughts
1. Single vs. Multiple paths
  * CoT uses 1 direction to get answers.
  * ToT uses multiple decision branches.

2. Sequenial Reasoning
  * CoT generates a sequence of thoughts.
  * ToT considered various sequences to derive best outcomes.

3. Flexibility in Approach
  * CoT is rigid with no room to change reasoning.
  * ToT can reverse or even change its reasoning paths if they dont actually lead to a solution.

## Tree of Thoughts Example
1. Input prompt
2. LLM generates 5 or more plans or answers and "votes" for the best answers.
3. Final top answer is given.

### Pros for Tree of Thoughts Prompting
1. Enables LLMs to plan and select from various problem-solving strategies for better outcomes.
2. Enahnces adaptability in problem-solving by allowing the model to **revise its decisions.**
3. Outperforms other methods by **evaluating multiple options for complex problems.**
4. Facilitates understanding of decision-making by **tracing models reasoning steps.**

### Cons for Tree of Thoughts Prompting
1. May consume more computing resources than simpler methods, increasing costs and token usage.
2. Unnecessary for simpler tasks where less complex methods suffice.
3. Considered "experimental" in some regards, may not be universally effective for all problem types.


## 3. Graph of Thoughts Prompting
* Method to help LLMs organize information in a graph network with each data point connected similar to neural pathways.
* Promotes collaborative data points thus enhancing solution generation and iterative problem-solving.

### Comparing Graph of Thoughts vs. Tree of Thoughts
* **Graph of Thoughts** creates a dynamic web of interconnected ideas.
* **Tree of Thoughts** develops a hierarchical structure of ideas.

Summary:
* The key differences:
1. Graph of Thoughts enables the synthesis of mixed/variable ideas.
2. Tree of Thoughts progresses along distinct non-converging paths similar to a decision tree algorithm.

### Pros for Graph of Thoughts Prompting
1. Fosters robust solutions by integrating divers ideas.
2. Stimulates innovateive thinking through idea interconnectivity.
3. Enhances task quality, improving efficiency and effectiveness.
4. Offers flexibility and evolves with addition of new ideas.

### Cons of Graph of Thought Prompting
1. Complexity may hinder the understanding and usability due to multiple idea connections.
2. Web of thoughts could make finding the optimal solution very challening!
3. Processing numerous connections may slow down the solution identification.
4. Begineers may struggle with optimizing the combination of improvement of thoughts.
